```python
data = pd.read_csv("../../data/model_data.csv")
```

```python
col =['이용개월수_일시불_R3M', '이용건수_체크_R3M', '이용금액_할부_R6M', '이용개월수_체크_R3M', '이용금액_할부_R3M', '이용건수_일시불_R12M', '이용금액_일시불_R12M', '이용건수_일시불_R3M', '이용건수_체크_R12M', '이용건수_할부_R12M', '이용건수_체크_R6M', '이용금액_체크_R3M', '이용건수_할부_R3M', '이용개월수_할부_R6M', '이용건수_일시불_R6M', '이용개월수_일시불_R6M', '이용금액_체크_R6M', '이용개월수_체크_R6M', '이용금액_일시불_R6M', '이용금액_일시불_R3M', '이용금액_할부_R12M', '이용금액_체크_R12M', '이용금액_R3M_신용체크', '탈회횟수_발급1년이내', '이용개월수_할부_R3M', '이용건수_할부_R6M','청구금액_B0']
data = data.drop(columns=col)
```

```python
new_data = new_data.drop(columns = '발급회원번호')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

# 수치형 컬럼만 추출
numeric_cols = new_data.select_dtypes(include='number').columns

plt.rcParams['font.family'] = 'Malgun Gothic'      # 한글 폰트 설정 (Windows용)
plt.rcParams['axes.unicode_minus'] = False         # 마이너스 깨짐 방지

for col in numeric_cols:
    # 숫자로 변환 + 결측치 제거
    series1 = pd.to_numeric(data[col], errors='coerce').dropna()
    series2 = pd.to_numeric(new_data[col], errors='coerce').dropna()

    # 값의 범위 설정
    min_val = min(series1.min(), series2.min())
    max_val = max(series1.max(), series2.max())
    x_vals = np.linspace(min_val, max_val, 1000)

    # KDE 계산
    kde1 = gaussian_kde(series1)
    kde2 = gaussian_kde(series2)
    y1 = kde1(x_vals)
    y2 = kde2(x_vals)

    # 겹치는 부분 계산
    overlap = np.minimum(y1, y2)

    # 그래프 그리기
    plt.figure(figsize=(8, 4))
    plt.fill_between(x_vals, 0, y1, color='blue', alpha=0.5, label='train_data')
    plt.fill_between(x_vals, 0, y2, color='orange', alpha=0.5, label='new_data')
    plt.fill_between(x_vals, 0, overlap, color='purple', alpha=0.7, label='겹치는 영역')

    plt.title(f'분포 비교: {col}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# 범주형 컬럼만 추출
categorical_cols = new_data.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    train_counts = data[col].value_counts(normalize=True)
    new_counts = new_data[col].value_counts(normalize=True)
    compare_df = pd.DataFrame({'train': train_counts, 'new': new_counts}).fillna(0)

    compare_df.plot(kind='bar')
    plt.title(f'범주형 변수 분포 비교: {col}')
    plt.ylabel('비율')
    plt.tight_layout()
    plt.show()
```

```python
missing_df = pd.DataFrame({
    'train_missing': data.isnull().mean(),
    'new_missing': new_data.isnull().mean()
})

missing_df.plot(kind='bar', figsize=(10, 4), title='결측치 비율 비교')
plt.tight_layout()
plt.show()
```
