```python
vip = pd.read_csv("../../data/vip_member_data2.csv")
```

```python
vip
```

```python
final_merge = pd.merge(vip, merged_df, on='발급회원번호', how='inner')
```

```python
target = pd.read_csv("../../data/청구금액.csv")
```

```python
target
```

```python
target_merge = pd.merge(final_merge, target , on = '발급회원번호', how = 'inner')
```

```python
target_merge
```

```python
target_merge['발급회원번호'] = target_merge['발급회원번호'].str.replace('SYN_','',regex=False).astype(int)
```

```python
target_merge['연령'] = target_merge['연령'].str.replace('대|이상','',regex=True).astype(int)
```

```python
target_merge = target_merge.drop(columns=['발급회원번호','청구금액_R3M','청구금액_R6M'])
```
